## Working With Missing Data

1. Finding Missing Data

```python
data = [
    {"Name": "Rahul", "Age": 25, "City": "Delhi", "Salary": 50000},
    {"Name": "Priya", "Age": None, "City": "Mumbai", "Salary": 62000},
    {"Name": "Amit", "Age": 30, "City": None, "Salary": 58000},
    {"Name": "Sneha", "Age": 28, "City": "Chennai", "Salary": None},
    {"Name": None, "Age": 35, "City": "Bangalore", "Salary": 72000},
    {"Name": "Ravi", "Age": None, "City": None, "Salary": 49000}
]
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()#Count the all Nan Values
```

```python
df.isna().any() #it will give Ture if atleast one value null in the data
```

2. Removing Missing Data

```python
#dropna(): remove Which row have atleast one null value
```

```python
#dropna(thresh=3):Will romeve which row does't have atleast 3 non-Null values
df.dropna(thresh=3)
```

3. Filling the missing data

```python

data = [
    {"Age": 25, "Salary": 50000, "Experience": 2},
    {"Age": 30, "Salary": None, "Experience": 5},
    {"Age": 28, "Salary": 58000, "Experience": None},
    {"Age": None, "Salary": 62000, "Experience": 4},
    {"Age": 35, "Salary": 72000, "Experience": 8},
    {"Age": None, "Salary": 49000, "Experience": None}
]

df1=pd.DataFrame(data)
df1
```

```python
df1.fillna(0)
df1
```

```python
#filling with mean,mode and median
df1.fillna(df1.mean())
```

```python
df1.fillna(df1.mode())
```

```python
df1.fillna(df1.median())
```

```python
df1[df1['Age']==25]
```

```python
df1.fillna(2,inplace=True)
```
