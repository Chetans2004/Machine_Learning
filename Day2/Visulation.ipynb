## 1.Relational Plots

Relational plots are used to visualize relationships between two or more variables — especially numerical ones.
They help you see correlations, patterns, clusters, or trends in your data.
- lineplot:To show how one variable changes with another (usually over time or in a continuous manner)
- scatterplot:To visualize the relationship between two numeric variables using dots.
- relplot:relplot() is a wrapper (general function) that can create either a line plot or scatter plot, depending on the parameter kind

```python
sns.lineplot( data=data,x='sepal_length',y='sepal_width')
```

```python
#hue;Each species (e.g., setosa, versicolor, virginica) gets its own color
sns.lineplot( data=data,x='sepal_length',y='sepal_width',hue='species')

```
