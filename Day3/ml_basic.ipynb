## Day 3 – Machine Learning Basics
What is Machine Learning?

Machine Learning (ML) is a branch of Artificial Intelligence (AI) that enables systems to learn automatically from data and improve from experience without being explicitly programmed.

The goal of ML is to make computers learn patterns from data and make predictions or decisions.
Real-world Examples:
Spam email detection
House price prediction
Netflix or YouTube recommendation system
Self-driving cars














## How Machine Learning Works

1. Collect Data – Gather raw data from various sources.
2. Preprocess Data – Clean, handle missing values, and prepare features.
3. Split Data – Divide data into training and testing sets.
4. Train Model – Teach the model using the training data.
5. Test Model – Evaluate performance on the test data.
6. Deploy Model – Use it for real-world predictions.


## Types of Machine Learning
1. Supervised Learning

Model learns from labeled data (data with known outputs).
The goal is to predict outcomes for new, unseen data.

 Examples:
Regression: Predicting continuous values (e.g., house prices).
Classification: Predicting categories (e.g., spam or not spam).

Algorithms:
Linear Regression
Logistic Regression
Decision Trees
Random Forest

2.  Unsupervised Learning

Model learns from unlabeled data — no predefined output.
It tries to find hidden patterns or structure in the data.

Examples:
Grouping customers based on behavior (clustering)
Reducing dimensions for visualization (PCA)

 Algorithms:
K-Means Clustering
Hierarchical Clustering
Principal Component Analysis (PCA)

3. Reinforcement Learning
Model (called an agent) learns by interacting with an environment.
It receives rewards or penalties based on actions taken.
The goal is to maximize cumulative rewards.

 Examples:

Playing games (e.g., Chess, Go, Atari)
Self-driving cars
Robotics

## Summary

ML helps computers learn and adapt from data.
- Three main types:
- Supervised Learning: Labeled data → Predict outputs
- Unsupervised Learning: Unlabeled data → Find patterns
- Reinforcement Learning: Learn by reward and punishment

This forms the foundation for building future models like regression, classification, and clustering.
