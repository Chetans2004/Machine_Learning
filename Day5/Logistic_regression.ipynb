```python

Q1 = df['Pregnancies'].quantile(0.25)
Q3 = df['Pregnancies'].quantile(0.75)

IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR


Pregnancies_outliers = df[(df['Pregnancies'] < lower_limit) | (df['Pregnancies'] > upper_limit)]
Pregnancies_outliers


#Pregnancies_outliers percentage
percentage_outliers_Pregnanciesr = (len(Pregnancies_outliers) / len(df)) * 100
percentage_outliers_Pregnanciesr
```

```python
#Since  outlier are below 5% we need replace
df.loc[df['Pregnancies']>upper_limit,'Pregnancies'] =df['Pregnancies'].median()
df.loc[df['Pregnancies']<lower_limit,'Pregnancies']=df['Pregnancies'].median()
```
