## Combining DataFrame

```python
df=pd.DataFrame({'A':['A0','A1','A2','A3','A4'],
                   'B':['B0','B1','B2','B3','4'],
                 'C':['C0','C1','C2','C3','C4'],
                 'D':['D0','D1','D2','D3','D4']})
df
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3','A4'],
                   'B':['B0','B1','B2','B3','4'],
                 'C':['C0','C1','C2','C3','C4'],
                 'D':['D0','D1','D2','D3','D4']})
df1
```

```python
pd.concat([df, df1], axis=0,ignore_index=True)#Axis 0 means row wise (one DataFrame below another)
```

```python
pd.concat([df, df1], axis=1)#Axis 1 means row wise (Side by side)
```
